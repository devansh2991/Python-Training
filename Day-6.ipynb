```python
class bankaccount:
  def __init__(self, balance):
    self.balance = balance
  def deposit(self, amount):
    self.balance += amount
    return self.balance
  def withdraw(self, amount):
    if self.balance >= amount:
      self.balance -= amount
      return self.balance
  def get_balance(self):
    return self.balance

acc = bankaccount(500)
print(acc.get_balance())
print(acc.deposit(500))
print(acc.withdraw(200))


import numpy as np

a = np.array([1,2,3,4])
b = np.zeros((2,3))
c = np.ones((2,3))
d = np.full((2,3), 7)
e = np.arange(0, 10, 2)
print(a,b,c,d,e)

a.shape[0]

reshaped = a.reshape(2,2)
print(reshaped)

arr = np.array([[10,20,30],[40,50,60],[70,80,90]])
print(arr[0:2, 1:3])

x = np.array([1,2,3])
y = np.array([4,5,6])
print(np.sqrt(x))
print(np.sum(x, axis=0))

rand = np.random.rand(3,3)
rand_ints = np.random.randint(0,10,(2,3))
normal_dist = np.random.normal(1,2,5)
print(rand)
print(rand_ints)
print(normal_dist)
```
